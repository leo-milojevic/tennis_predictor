## Plot accuracy of prediction at each percent

```python
outcomes['result'] = outcomes[0].apply(lambda x: round(x, 2))
accuracy = {}

for odds in np.arange(0.5, 1, 0.01):
    acc = len(outcomes[(outcomes[0] > odds) & (outcomes[0] < odds+0.01)]) # all the times that the player with the odds won
    inacc = len(outcomes[(outcomes[0] < 1 - odds) & (outcomes[0] > 0.99 - odds)]) # all the times that the player with the odds lost
    accuracy[odds] = acc/(acc+inacc)
```

```python
pd.DataFrame.from_dict(accuracy,orient='index').plot(figsize=(12,8))
```

```python
from sklearn.metrics import mean_squared_error

print('RMSE(real, accuracy): ' + str(mean_squared_error(np.arange(0.5, 1, 0.01), list(accuracy.values())) ** 0.5))
```
